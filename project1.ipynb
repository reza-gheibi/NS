```python
# Required Header
%matplotlib inline

import matplotlib.pyplot as plt
import networkx as nx
```

## Get a edge list data set
### 1. Download a edge list that has more than 1000 nodes. Careful not to pick very large graphs
### 2. Store the edge list as a graph call it it G1

```python
el = open ("facebook_combined.txt", 'rb') # open file and rb is for read-binary
G1 = nx.read_edgelist(el, nodetype=int) # edges on each line of the input
el.close() # we can close the file now
```

## Plot the graph G1 without node numbers

```python
#nx.draw_circular(G1, node_color='C0',node_size=10, width=0.01, with_labels=False)
nx.draw_networkx(G1, node_size=5, width=0.09, with_labels=False)
plt.show(G1)
```

## Find the degree of each node in G1 and draw a histogram: nodes on the x-axis and degree on the y-axis

```python
#degree_sequence = [d for n, d in G1.degree] #creating the degree sequence with nodes and degrees
#degreeHist = {} #creating the empty dic to save histogram
#for d in degree_sequence:
#    if d in degreeHist:
#        degreeHist[d] += 1 
#    else:
#        degreeHist[d] = 1
#n = degreeHist.values()  #the values in degreeHist are our number of nodes with a specific degree      
#d = degreeHist.keys()    #the keys in degreeHist are the unique degrees 
#plt.bar(n, d, width=0.5, color="b")  #ploting the histogram

n = [node for (node, val) in G1.degree()] #storing the nodes
d = [val for (node, val) in G1.degree()]  #storing the degrees
plt.bar(n, d, width=0.5, color="b")  #ploting the histogra
plt.title("Degree Histogram")
plt.ylabel("degree")
plt.xlabel("node")
plt.show()

#in graph G1 we have 4038 nodes and maximum degree is 1045
```

## Find the average degree of nodes in G1

```python
degree_sequence = [d for n, d in G1.degree()] 
average = sum(degree_sequence) / len(degree_sequence) #get the summation of all degrees and divide it by number of them(nodes)
avg=round(average) # since we do not have a degree with a float number we round it
print("The rounded average degree of nodes in G1 = " + str(avg))
```

## Construct a subgraph G2 from G1 by removing nodes whose degree is less than 2 times the average degree and nodes whose degree is greater than 2 times the avearge degree

```python
#new_nodes =  sorted(G1.degree, key=lambda x: x[0])
#new_nodes= degree_sequence
s=sorted(G1.degree)
result_1 = [i for i in s if i[1] == avg*2] #exaclty equal to average

#result_2 = [i for i in s if i[1] < avg*2]
#result_3 = [i for i in s if i[1] > avg*2]

f1=[i[0] for i in result_1] # exactly equal to average

#f2=[i[0] for i in result_2]
#f3=[i[0] for i in result_3]

G2=G1.subgraph(f1) # exactly equal to average

#G3=G1.subgraph(f2)
#G4=G1.subgraph(f3)

#nx.draw_circular(G2, node_color='C0', node_size=50, with_labels=False)
#plt.show(G2)
#nx.draw_circular(G3, node_color='C0', node_size=50, with_labels=False)
#plt.show(G3)
```
